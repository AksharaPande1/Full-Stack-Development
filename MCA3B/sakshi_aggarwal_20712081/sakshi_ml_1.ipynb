```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("pollution.csv")
df.head()
```

```python
import seaborn as sns
sns.barplot(x="NO2μg/l", y="Air Quality",data=df)
plt.show()
```

```python
plt.hist(df["Air Quality"])
plt.xlabel('NH3 Air')
plt.show()
```

```python
sns.displot(df["Air Quality"])
```

```python
sns.scatterplot(x="NO2μg/l", y="Air Quality", data=df)
plt.title('Air Quality Of Dehradun')
plt.show()
```
